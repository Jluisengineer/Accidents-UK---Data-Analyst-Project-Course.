#### Vehicle.csv

This notebook is used to clean the file vehicles 5 years.csv

We will get three files at the end:
* cars.csv
* motorbikes.csv
* Accident_conditions.csv

```python
# Libraries I need

import pandas as pd
import os
```

```python
# This path is in my computer. It wo't work in your computer. You have to replace it with your path
mainpath ='C:/Users/jluis/Desktop/Programacion/Programacion - Data analyst/Python/Machine learning/data/'
filename = 'Accidents 5 years/Vehicle 5 years.csv'
fullname = os.path.join(mainpath, filename)
# Reading the DataFrame
df = pd.read_csv(fullname, header=0,delimiter=',')
```

```python
df.head()
```

```python
df.columns.values
```

```python
df = df[["sex_of_driver","age_of_driver","vehicle_type","generic_make_model","engine_capacity_cc", "age_of_vehicle"]]
df
```

```python
# Dataframe with cars
cars = df[(df["vehicle_type"]==8) | (df["vehicle_type"]==9)]
cars
```

```python
# Dataframe with motorbikes
motorbikes = df[(df["vehicle_type"]==2) | (df["vehicle_type"]==3) | (df["vehicle_type"]==4) | (df["vehicle_type"]==5)]
motorbikes
```

```python
# Exporting clean data sep = "," (default)
cars.to_csv("cars.csv",encoding="utf-8")
motorbikes.to_csv("motorbikes.csv", encoding="utf-8")
```
