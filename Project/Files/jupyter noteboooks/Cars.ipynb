#### Engine capacity

```python
cars = cars[['sex_of_driver', 'age_of_driver', 'vehicle_type','generic_make_model', 'engine_capacity_cc', 'age_of_vehicle']]
```

```python
cars.head()
```

```python
engine_car = cars[(cars["engine_capacity_cc"]>=454) & (cars["engine_capacity_cc"]<2874)]
engine_car
```

```python
plt.hist(engine_car["engine_capacity_cc"])
plt.title(("Histogram: Cars Engine Capacity"))
plt.xlabel("Cars: Engine capacity")
plt.ylabel("Cars: Frequency")
```

```python
engine_car["engine_capacity_cc"].describe()
```

```python
data = pd.crosstab(index=engine_car['engine_capacity_cc'], columns='frequency')
data = data[data["frequency"]>500]
```

```python
plt.scatter(data.index, data["frequency"])
plt.title("Relation between Accidents and Engine capacity")
plt.xlabel("Cars: Engine capacity CC")
plt.ylabel("Cars: Accident Frequency")
```
